# Very Dense Intersection: Reward-Suited Creeping vs Non-Suited Reward

This notebook repeats the straight and left-turn HighwayEnv intersection experiment in a very dense benchmark:

- `initial_vehicle_count = 10`
- `spawn_probability = 0.60`
- route-following ego with learned longitudinal attention policy
- negotiating traffic enabled: cross traffic yields only when the ego sends a creeping/slowdown signal

I interpret the requested ?90% collision rate? as the prior collision-avoidance target: at least 90% success / at most 10% collision for the reward-suited creeping policy.

```python
from pathlib import Path
import pandas as pd

RESULTS = Path("notebooks/results/intersectionRouteAttentionCreeping")
VIDEOS = Path("notebooks/results/route_videos/very_dense_final_suited_policy_10eps_20260614-215712")
summary = pd.read_csv(RESULTS / "very_dense_final_reward_suited_vs_not_summary.csv")
cols = ["maneuver", "agent", "episodes", "success_rate", "collision_rate", "mean_creep_zone_speed", "high_speed_zone_rate", "creep_speed_rate", "mean_min_conflict_ttc"]
summary[cols]
```

## Result

The reward-suited creeping policies clear the target in the very dense benchmark:

| Maneuver | Reward / policy | Success | Collision | Mean creep-zone speed | High-speed zone rate | Creep-speed rate |
|---|---:|---:|---:|---:|---:|---:|
| Straight | non-suited fast reward | 54.2% | 45.8% | 8.91 m/s | 99.2% | 0.0% |
| Straight | creeping-suited attention policy | 90.8% | 9.2% | 4.11 m/s | 1.6% | 88.9% |
| Left turn | non-suited fast reward | 51.7% | 48.3% | 8.90 m/s | 99.2% | 0.0% |
| Left turn | creeping-suited attention policy | 94.2% | 5.8% | 5.27 m/s | 40.1% | 54.0% |

The straight policy is cleanly creeping. The left policy meets the success/collision target and is substantially slower than the non-suited reward, though it still commits faster in some left-turn episodes.
